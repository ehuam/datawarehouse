### DECOMPOSE ONE WEBPAGE

```
webpage_path = webpages_folder / 'finviz_etf_page_one.html'
if not webpage_path.exists():
    try:
        driver = webdriver.Firefox(
                                    service=FirefoxService(GeckoDriverManager().install()),
                                    options=options
        )
        driver.get(FINVIZ_ETF_PAGE_ONE_URL)
        
        logger.info('Explicit wait for page to load')
        wait = WebDriverWait(driver, 15)
        wait.until(EC.presence_of_element_located((By.CLASS_NAME, "styled-row")))
        
        with open(webpage_path, 'w', encoding='utf-8') as f:
            f.write(driver.page_source)
        logger.info(f"{FINVIZ_ETF_PAGE_ONE_URL=} saved to {webpage_path}")
        
    except Exception as e:
        logger.error(f"error fetching {FINVIZ_ETF_PAGE_ONE_URL=}: {e}")
    
    finally:
        driver.quit()
        logger.info("Selenium WebDriver closed")
logger.info(f"{webpage_path} already exists. Skipping download.")
    
```

```
TAG_MAPPING = {
    'etf_ticker': 'data-boxover-ticker',
}

SITE_INVENTORY = {
    "control_plane": ["screener-combo-title", "screener-combo-select"],
    "navigation_tabs": ["Descriptiv", "ExchangeAn", "OverviewVa"],
    "data_rows": ["styled-row"],
    "layout_noise": ["header", "navbar", "footer", "modal-elite-ad", "script", "noscript", "iframe",
                     "js-elite-features-root", "notifications-container", "notifications-react-root", "dialogs-react-root",
                     "root", "IC_D_1x1_1", "portal/_r_5_", "ICUid_Iframe", "img", "svg", "use", "js-feature-discovery-root",
                     "screener-presets-root"
                     ],
    "navigation_controls": ["pageSelect", "pages-combo", "is-next", "screener-pages"]
}

# for performance
INV_HASH = {item: category for category, items in SITE_INVENTORY.items() for item in items}
```

```
# areas of interest overlay

FUNCTIONAL_AREAS = {
    "SCREENER_FILTERS": "filter-table-top",
    "DATA_TABLE": "screener-views-table",
    "PAGINATION_DROP": "pageSelect", # the drop down 
    "PAGINATION_NAV": "screener_pagination" # 1 2 3 ... 100 ... END -> on the page
}

dom_tree_data = {
    "name": "ETF_SCREENER_AGGREGATE",
    "data": {"tag": "synthetic", "semantic_role": "page_root"},
    "children": []
}
```

```
if webpage_path.exists():
    html_blob = webpage_path.read_text(encoding='utf-8')
    soup = BeautifulSoup(html_blob, 'html.parser')
        
    for label, element_id in FUNCTIONAL_AREAS.items():
        found_node = soup.find(id=element_id)
        if found_node:
            branch = aux_funcs.build_tree_data(
                found_node,
                max_depth=12,
                inventory_hash=INV_HASH
            )
            if branch:
                branch["name"] = f"{label} | {branch['name']}"
                dom_tree_data["children"].append(branch)
    
    logger.info(f"mapped DOM tree starting from {soup.body.name}")
 
```
