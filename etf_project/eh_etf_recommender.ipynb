# LOAD ENVIRONMENT VARIABLES

```
# DESCRIBE FOLDERS USED
current_working_dir = Path.cwd()
env_path = current_working_dir / 'ENV_PUBLIC'

# HTML DOWNLOADS FOLDER
webpages_folder = current_working_dir / 'webpages'
webpages_folder.mkdir(exist_ok=True)
```

```
load_dotenv(env_path)

# construct URI for MongoDB Connection
URI = (
    f"mongodb://{os.getenv('MONGO_INITDB_ROOT_USERNAME')}:"
    f"{os.getenv('MONGO_INITDB_ROOT_PASSWORD')}@"
    f"{os.getenv('MONGO_HOST')}:"
    f"{os.getenv('MONGO_PORT')}/"
)
CLIENT = MongoClient(URI, serverSelectionTimeoutMS=5000)
DB_NAME = os.getenv('MONGO_DB_NAME')
DB = CLIENT[DB_NAME]

```

### DECOMPOSE ONE WEBPAGE

```
fin_viz_landing_page_html = webpages_folder / 'finviz_etf_page_one.html'
if fin_viz_landing_page_html.exists():
    logger.info(f"File {fin_viz_landing_page_html} already exists. Skipping download.")
else:
    response = requests.get(FINVIZ_ETF_PAGE_ONE_URL)
    with open(fin_viz_landing_page_html, 'w', encoding='utf-8') as f:
        f.write(response.text)
    logger.info("Saved landing ETF page from Finviz to local directory.")
```
