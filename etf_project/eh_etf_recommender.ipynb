```
#IMPORTS
import sys
from pymongo import MongoClient
from dotenv import load_dotenv
from pathlib import Path
import os
import logging
from random_user_agent.user_agent import UserAgent
from random_user_agent.params import SoftwareName, OperatingSystem
from bs4 import BeautifulSoup
import requests
```

```
logging.basicConfig(
    level=logging.INFO,
    format='%(name)s - %(levelname)s - %(asctime)s - %(message)s',
)
logger = logging.getLogger(__name__)
```

# LOAD ENVIRONMENT VARIABLES

```
current_working_dir = Path.cwd()
env_path = current_working_dir / 'ENV_PUBLIC'

load_dotenv(env_path)

# construct URI for MongoDB Connection
URI = (
    f"mongodb://{os.getenv('MONGO_INITDB_ROOT_USERNAME')}:"
    f"{os.getenv('MONGO_INITDB_ROOT_PASSWORD')}@"
    f"{os.getenv('MONGO_HOST')}:"
    f"{os.getenv('MONGO_PORT')}/"
)
CLIENT = MongoClient(URI, serverSelectionTimeoutMS=5000)
DB_NAME = os.getenv('MONGO_DB_NAME')
DB = CLIENT[DB_NAME]

```

```
# CHECK CONNECTION TO MONGODB CONTAINER
try:
    CLIENT.admin.command('ping')
    logger.info("Sucessfully connected to mongoDB")
except Exception as e:
    logger.warning(f"ERROR: Could not connect to MongoDB. \nDetails: {e}")
    logger.warning("Troubleshooting steps: \n \
                   1. Ensure Colima/Docker is running. \n \
                   2. Verify ENV_PUBLIC file contains MONGO_HOST \n \
                   3. Run: docker-compose --env-file ENV_PUBLIC up -d")
    sys.exit(1)
```
